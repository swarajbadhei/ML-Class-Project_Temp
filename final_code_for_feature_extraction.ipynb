```python
import sys
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import librosa
```

```python
#base variables
total_processed=0
total_failed=0
target_values=[]
#auxiliary variables
total_ambulance_processsed=0
total_ambulance_failed=0
total_ambulance_size_issue=0
arr_zr=np.zeros(shape=(1,40),dtype=float)
for subdir, dirs,files in os.walk('drive/My Drive/usha maam prjct/audio_data/ambulance'):
  for filename in files:
    filepath=subdir+os.sep+filename
    if filepath.endswith('webm'):
      try:
        data,sampling_rate=librosa.load(filepath)
        mfcc=librosa.feature.mfcc(y=data,sr=sampling_rate,n_mfcc=40)
        #mfcc=mfcc.T
        if mfcc.shape[1]==861:
          #dont touch this thing
          arr_zr=np.vstack((arr_zr,mfcc.T))
          total_processed+=1
          total_ambulance_processsed+=1
          print('done for file:{}'.format(filepath))
          target_values.append(0)
        else:
          print('shape issue for file : {}'.format(filepath))
          total_ambulance_size_issue+=1
          pass
      except:
        total_failed+=1
        total_ambulance_failed+=1
        print("Failed for file:{}".format(filepath))
        pass
```

```python
print('Total ambulance processed = {}'.format(total_ambulance_processsed))
print('Total ambulance failed = {}'.format(total_ambulance_failed))
print('Total ambulance size issue = {}'.format(total_ambulance_size_issue))
print(target_values)
```

```python
print('Total Fire processed = {}'.format(total_fire_processed))
print('Total fire failed = {}'.format(total_fire_failed))
print('Total fire size issue = {}'.format(total_fire_size_issue))
print(target_values)
```
